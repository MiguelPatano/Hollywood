1997: Estreno de "Titanic", que se convirtió en la película más taquillera hasta ese momento.
2009: Estreno de "Avatar", otra revolución en los efectos especiales y 3D.
2010s en adelante: Crecimiento de las plataformas de streaming y cambios en la forma de distribución y consumo de cine.

```python
#Creo la columna de Eras del cine y la de décadas
def categorize_era(year):
    if year >= 1890 and year < 1930:
        return 'Era del Cine Mudo'
    elif year >= 1930 and year < 1950:
        return 'Era del Cine Clásico de Hollywood'
    elif year >= 1950 and year < 1970:
        return 'Era de la Postguerra y la Edad de Oro del Cine'
    elif year >= 1970 and year < 1990:
        return 'Era del Nuevo Hollywood y el Cine Moderno'
    elif year >= 1990:
        return 'Era del Cine Contemporáneo y la Globalización'
    else:
        return 'Desconocida'

def categorize_decade(year):
    return (year // 10) * 10

# Añadir columna de grandes grupos de tiempo (eras del cine)
df_movies['era'] = df_movies['año'].apply(categorize_era)

# Añadir columna de décadas
df_movies['decade'] = df_movies['año'].apply(categorize_decade)
```

```python
#Voy a calcular calificación ponderada para las películas, usando como ejemplo la que usa IMDB
# Weighted Rating (WR) = (v / (v + m) * R) + (m / (v + m) * C)
# Donde: R es el promedio de la calificación de la película
#       v es el número de votos de la película 'vote_count' para la base original o numVotes que es de la otra base
#       m es el umbral mínimo de votos necesarios para ser considerado vamos a considerar el percentil 50
#       C es el promedio general de todas las calificaciones 'vote_average' o 'averageRating'
 
 
# Vamos a aumentar el Numvotes de todas las películas situadas como outliers por debajo del q1 para que no molesten en el cálculo
 
# Calculamos los cuartiles para NumVotes
Q1 = df_merged['numVotes'].quantile(0.25)
Q3 = df_merged['numVotes'].quantile(0.75)
IQR = Q3 - Q1
 
# Definimos el umbral para identificar outliers en NumVotes
lower_bound = Q1 - 1.5 * IQR
 
# Filtramos las películas que son outliers en NumVotes
outliers = df_merged[df_merged['numVotes'] < lower_bound]
 
# Ajustamos las averagerating de las películas outliers para que estén en el primer cuartil de averagerating
Q1_rating = df_merged['averageRating'].quantile(0.25)
outliers['averageRating'] = Q1_rating
 
 
# Reemplazamos las películas outliers en el DataFrame original
df_merged.loc[outliers.index, 'averageRating'] = Q1_rating
 
# Calcular C
C = df_merged['averageRating'].mean()
 
# Definimos m en percentil 50
m = df_merged['numVotes'].quantile(0.50)
# Pongo el umbral mínimo de votos para no penalizar
umbral_minimo = 1000
 
# Función Weighted Rating (WR) asignamos un valor incluso si los votos son bajos
def weighted_rating(x, m=m, C=C):
    v = x['numVotes']
    R = x['averageRating']
    if v >= m:
        return (v / (v + m) * R) + (m / (v + m) * C)
    else:
        # Para películas con menos votos aplico una penalización
        penalizacion = v / umbral_minimo
        return min((R * penalizacion) + (C * (1 - penalizacion)), 6)
 
# Aplicar la función a todas las películas
df_merged['weightedRating'] = df_merged.apply(weighted_rating, axis=1)
df_merged['weightedRating'] = df_merged.apply(
    lambda row: row['averageRating'] if row['weightedRating'] < 0 else row['weightedRating'], axis=1)
 
# #####ORIGINAL
# # Vamos a calcular C
# C = df_merged['averageRating'].mean()
 
# # Definimos m en  percentil 90
# m = df_merged['numVotes'].quantile(0.40)
 
# # Filtrar las películas que tienen al menos m votos
# qualified_movies = df_merged[df_merged['numVotes'] >= m]
 
# # Función Weighted Rating (WR)
# def weighted_rating(x, m=m, C=C):
#     v = x['numVotes']
#     R = x['averageRating']
#     return (v / (v + m) * R) + (m / (v + m) * C)
 
# # Aplicar la función a las películas calificadas
# qualified_movies['weightedRating'] = qualified_movies.apply(weighted_rating, axis=1)
# df_merged = pd.merge(df_merged, qualified_movies[['imdb_id', 'weightedRating']], on='imdb_id', how='left')
```

```python
#Vamos a preparar la base para el sistema de recomendación.

import string
import nltk
import spacy
from nltk.corpus import stopwords



# Descargar recursos necesarios de NLTK y SpaCy
nltk.download('stopwords')
nltk.download('punkt')
nlp = spacy.load('en_core_web_sm')



# Manejar los valores NaN
df_merged['runtime_level'] = df_merged['runtime_level'].astype(str)
df_merged['runtime_level'] = df_merged['runtime_level'].astype(int)
df_merged.fillna('', inplace=True)
# Función para procesar texto
def preprocess_text(text):
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    tokens = nltk.word_tokenize(text)
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]
    tokens = [nlp(word)[0].lemma_ for word in tokens]
    return ' '.join(tokens)

# Aplicar el procesamiento de texto a 'tagline' y 'keywords'
df_merged['tagline'] = df_merged['tagline'].apply(preprocess_text)
df_merged['keywords'] = df_merged['keywords'].apply(preprocess_text)
df_merged['title'] = df_merged['title'].str.lower()
```

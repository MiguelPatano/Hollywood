```python
#Creo la columna de Eras del cine y la de dÃ©cadas
def categorize_era(year):
    if year >= 1890 and year < 1930:
        return 'Era del Cine Mudo'
    elif year >= 1930 and year < 1950:
        return 'Era del Cine ClÃ¡sico de Hollywood'
    elif year >= 1950 and year < 1970:
        return 'Era de la Postguerra y la Edad de Oro del Cine'
    elif year >= 1970 and year < 1990:
        return 'Era del Nuevo Hollywood y el Cine Moderno'
    elif year >= 1990:
        return 'Era del Cine ContemporÃ¡neo y la GlobalizaciÃ³n'
    else:
        return 'Desconocida'

def categorize_decade(year):
    return (year // 10) * 10

# AÃ±adir columna de grandes grupos de tiempo (eras del cine)
df_movies['era'] = df_movies['aÃ±o'].apply(categorize_era)

# AÃ±adir columna de dÃ©cadas
df_movies['decade'] = df_movies['aÃ±o'].apply(categorize_decade)
```

```python
#Voy a calcular calificaciÃ³n ponderada para las pelÃ­culas, usando como ejemplo la que usa IMDB
# Weighted Rating (WR) = (v / (v + m) * R) + (m / (v + m) * C)
# Donde: R es el promedio de la calificaciÃ³n de la pelÃ­cula
#       v es el nÃºmero de votos de la pelÃ­cula 'vote_count' para la base original o numVotes que es de la otra base
#       m es el umbral mÃ­nimo de votos necesarios para ser considerado vamos a considerar el percentil 50
#       C es el promedio general de todas las calificaciones 'vote_average' o 'averageRating'
 
 
# Vamos a aumentar el Numvotes de todas las pelÃ­culas situadas como outliers por debajo del q1 para que no molesten en el cÃ¡lculo
 
# Calculamos los cuartiles para NumVotes
Q1 = df_merged['numVotes'].quantile(0.25)
Q3 = df_merged['numVotes'].quantile(0.75)
IQR = Q3 - Q1
 
# Definimos el umbral para identificar outliers en NumVotes
lower_bound = Q1 - 1.5 * IQR
 
# Filtramos las pelÃ­culas que son outliers en NumVotes
outliers = df_merged[df_merged['numVotes'] < lower_bound]
 
# Ajustamos las averagerating de las pelÃ­culas outliers para que estÃ©n en el primer cuartil de averagerating
Q1_rating = df_merged['averageRating'].quantile(0.25)
outliers['averageRating'] = Q1_rating
 
 
# Reemplazamos las pelÃ­culas outliers en el DataFrame original
df_merged.loc[outliers.index, 'averageRating'] = Q1_rating
 
# Calcular C
C = df_merged['averageRating'].mean()
 
# Definimos m en percentil 50
m = df_merged['numVotes'].quantile(0.50)
# Pongo el umbral mÃ­nimo de votos para no penalizar
umbral_minimo = 1000
 
# FunciÃ³n Weighted Rating (WR) asignamos un valor incluso si los votos son bajos
def weighted_rating(x, m=m, C=C):
    v = x['numVotes']
    R = x['averageRating']
    if v >= m:
        return (v / (v + m) * R) + (m / (v + m) * C)
    else:
        # Para pelÃ­culas con menos votos aplico una penalizaciÃ³n
        penalizacion = v / umbral_minimo
        return min((R * penalizacion) + (C * (1 - penalizacion)), 6)
 
# Aplicar la funciÃ³n a todas las pelÃ­culas
df_merged['weightedRating'] = df_merged.apply(weighted_rating, axis=1)
df_merged['weightedRating'] = df_merged.apply(
    lambda row: row['averageRating'] if row['weightedRating'] < 0 else row['weightedRating'], axis=1)
 
# #####ORIGINAL
# # Vamos a calcular C
# C = df_merged['averageRating'].mean()
 
# # Definimos m en  percentil 90
# m = df_merged['numVotes'].quantile(0.40)
 
# # Filtrar las pelÃ­culas que tienen al menos m votos
# qualified_movies = df_merged[df_merged['numVotes'] >= m]
 
# # FunciÃ³n Weighted Rating (WR)
# def weighted_rating(x, m=m, C=C):
#     v = x['numVotes']
#     R = x['averageRating']
#     return (v / (v + m) * R) + (m / (v + m) * C)
 
# # Aplicar la funciÃ³n a las pelÃ­culas calificadas
# qualified_movies['weightedRating'] = qualified_movies.apply(weighted_rating, axis=1)
# df_merged = pd.merge(df_merged, qualified_movies[['imdb_id', 'weightedRating']], on='imdb_id', how='left')
```

```python
#Vamos a preparar la base para el sistema de recomendaciÃ³n.

import string
import nltk
import spacy
from nltk.corpus import stopwords



# Descargar recursos necesarios de NLTK y SpaCy
nltk.download('stopwords')
nltk.download('punkt')
nlp = spacy.load('en_core_web_sm')



# Manejar los valores NaN
df_merged['runtime_level'] = df_merged['runtime_level'].astype(str)
df_merged['runtime_level'] = df_merged['runtime_level'].astype(int)
df_merged.fillna('', inplace=True)
# FunciÃ³n para procesar texto
def preprocess_text(text):
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    tokens = nltk.word_tokenize(text)
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]
    tokens = [nlp(word)[0].lemma_ for word in tokens]
    return ' '.join(tokens)

# Aplicar el procesamiento de texto a 'tagline' y 'keywords'
df_merged['tagline'] = df_merged['tagline'].apply(preprocess_text)
df_merged['keywords'] = df_merged['keywords'].apply(preprocess_text)
df_merged['title'] = df_merged['title'].str.lower()
```
